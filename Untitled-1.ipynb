```python
import gymnasium as gym
from stable_baselines3 import DQN
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.callbacks import BaseCallback

class MountainCarRewardWrapper(gym.RewardWrapper):
    def reward(self,reward):
        pos=self.env.unwrapped.state[0]
        vel=self.env.unwrapped.state[1]

        if pos>=0.45:
            reward+=100
        elif pos>=0.3:
            reward+=10
        elif pos>0.1:
            reward+=3
        reward+=vel*5
        return reward
    
class episodelogger(BaseCallback):
    def __init__(self):
        super().__init__()
        self.ep=0
    def _on_step(self):
        if self.locals["dones"][0]:
            self.ep+=1
            r=self.locals["infos"][0].get("episode",{}).get("r",0)
            print(f"{self.ep} {r}")
        return True

def create_env(env_name:str="MountainCar-v0",render:bool=False,use_shaping:bool=True):
    if render:
        env=gym.make(env_name,render_mode='human')
    else:
        env=gym.make(env_name)
    if use_shaping and not render:
        env=MountainCarRewardWrapper(env)
    env=Monitor(env)
    return env

def create_model(env:gym.Env)->DQN:
    model=DQN(
        policy='MlpPolicy',
        env=env,
        learning_rate=5e-4,
        gamma=0.99,
        buffer_size=100_000,
        learning_starts=1_000,
        batch_size=128,
        train_freq=4,
        target_update_interval=1_000,
        exploration_fraction=0.5,
        exploration_initial_eps=1.0,
        exploration_final_eps=0.05,
        verbose=0,
    )
    return model

def train_model(model:DQN,total_timesteps:int=200_000):
    model.learn(total_timesteps=total_timesteps,callback=episodelogger(),log_interval=True)
    model.save("dqn_mountaincar_model")

timesteps=500_000
env=create_env(use_shaping=True)
model=create_model(env)
train_model(model,total_timesteps=timesteps) 
```

```python
def demo_agent(env_name="MountainCar-v0", n_episodes: int = 20):
    total_rewards = []
    env = create_env(env_name, render=True)  
    model = DQN.load("dqn_mountaincar_model", env=env)
    
    for ep in range(n_episodes):
        obs, _ = env.reset()
        episode_reward = 0
        done = False
        
        while not done:
            action, _ = model.predict(obs, deterministic=True)
            obs, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            episode_reward += reward
        
        print(episode_reward)
    env.close()

demo_agent(n_episodes=5)

```
